```python
#factorial
def factorial(n):
  if(n==0):
    return 1
  else:
    return n*factorial(n-1)
```

```python
print(factorial(6))
```

```python
print(factorial(5))
```

```python
print(factorial(0))
```

```python
#fibonacci series
def fibonacci(n):
  if n==0:
    return 0
  if n==1:
    return 1
  else:
    return fibonacci(n-1)+fibonacci(n-2)
```

```python
fibonacci(5)
```

```python
fibonacci(6)
```

```python
fibonacci(10)
```

```python
# sum of the element of sn array
def sum_of(arr, n):
  if n<1:
    return 0
  else:
    return sum_of(arr,n-1)+arr[n-1]
arr = [1,2,3,4,5]
```

```python
print(sum_of(arr,len(arr)))
```

```python
#print the number of 1 to n
def print_number(n):
  #base condition
  if n>0:
    print_number(n-1)
    print(n)
```

```python
print_number(6)
```

```python
# print number n to 1
def print_number_reverse(n):
  if n>0:
    print(n)
    print_number_reverse(n-1)
```

```python
print_number_reverse(6)
```

```python
#climmbing the stairs
def climb_stairs(n):
  if(n==0):
    return 1
  elif n<0:
    return 0
  else:
    return climb_stairs(n-1)+climb_stairs(n-2)
```

```python
n=5
climb_stairs(n)
```

```python
# binary search using recursion
def binary_search(arr,left,right,target):
  if left>right:
    return -1
  mid = left+(right - left)//2
  if arr[mid]== target:
    return mid
  elif arr[mid]>target:
    return  binary_search(arr,left,mid-1,target)
  else:
    return binary_search(arr,mid+1,right,target)
```

```python
arr = [2,3,4,10,40]
target = 10
print(binary_search(arr,0,len(arr)-1,target))
```
