```python
def lower_bound_optimal(arr,x):
  low = 0
  high = len(arr)-1
  ans = len(arr)
  while low <= high:
    mid = (low+high)//2
    if arr[mid]>=x:
      ans = mid
      high = mid-1
    else :
      low = mid+1
  return ans

```

```python
a = [2, 2, 2, 4, 4, 6]
x = 0
lower_bound_optimal(a,x)
```

```python
x = 4
lower_bound_optimal(a,x)
```
