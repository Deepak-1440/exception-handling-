```python
def bubble_sort(arr):
  n = len(arr)
  for i in range (n-1):
    for j in range (n-i-1):
      if(arr[j]>arr[j+1]):
        arr[j], arr[j+1] = arr[j+1], arr[j]
```

```python
arr = [5,3,4,7,1]
print(arr)
bubble_sort(arr)
print(arr)
```
